```python
def get_values(df, field, max_values=20):

    if data[field].value_counts().shape[0] < max_values:
        display(pd.DataFrame(data[field].value_counts().reset_index()))
    else:
        print(field, ': ', data[field].value_counts().shape[0], ' values' )
    print('\n')
```
